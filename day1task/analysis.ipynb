```python
import matplotlib.pyplot as plt

missing = df.isnull().sum()

plt.figure(figsize=(10,5))
missing.plot(kind='bar')
plt.title('Missing Values per Column')
plt.xlabel('Columns')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
